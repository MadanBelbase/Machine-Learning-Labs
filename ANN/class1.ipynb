```python
class Perceptron:
    """A single-layer perceptron with the classic step-function learning rule."""

    def __init__(self, learning_rate=0.1, n_iters=20):
        self.lr = learning_rate
        self.n_iters = n_iters
        self.weights = None
        self.bias = None
        self.errors_ = []      # misclassifications per epoch (for the learning curve)

    def _activation(self, z):
        """Heaviside step function."""
        return np.where(z >= 0, 1, 0)

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0.0
        self.errors_ = []

        for _ in range(self.n_iters):
            errors = 0
            for idx in range(n_samples):
                # 1) forward pass
                z = np.dot(X[idx], self.weights) + self.bias
                y_pred = self._activation(z)

                # 2) perceptron learning rule
                update = self.lr * (y[idx] - y_pred)
                self.weights += update * X[idx]
                self.bias += update

                errors += int(update != 0.0)
            self.errors_.append(errors)
        return self

    def predict(self, X):
        z = np.dot(X, self.weights) + self.bias
        return self._activation(z)
```
