```python
# dataframe 
import pandas as pd
data = {
    "Year of Experience":[4,5,6,3],
    "Education": ["High School","Bachelor","High school","Master"],
    "Salary":[30000,45000,40000,65000]
}
df = pd.DataFrame(data)
print(df)
df.head()
df.info()
df.describe()
```

```python
# converting object to number 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Education'] = le.fit_transform(df['Education'])
df
```

```python
X = df[['Year of Experience','Education']]
y = df['Salary']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model =  LinearRegression()
```

```python
model.fit(X,y)
```

```python
c = model.intercept_
m = model.coef_
```

```python
print(c,m)
```

```python
print(f"y = {m[0]:.2f}x+{c:.2f}")
print(f"y = {m[1]:.2f}x+{c:.2f}")
```

```python
y_pred = model.predict(X)
```

```python
import numpy as np
df['Predict'] = np.round(y_pred, 2)
```

```python
df
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
Mse = mean_squared_error(y,y_pred)
r2 = r2_score(y,y_pred)
print("MSE = " ,Mse)
print("R2 = ", r2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y,y_pred)
plt.plot([y.min(),y.max()],[y.min(),y.max()])
plt.show()
```
