```python
# dataframe 
import pandas as pd
data = {
    "Area ": [50,70,75,85,90,101],
    "Price":[200,280,325,390,378,385]
}
df = pd.DataFrame(data)
print(df)
df.head()
df.info()
df.describe()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model =  LinearRegression()
```

```python
x = df[['Area ']]
y = df['Price']
```

```python
model.fit(x,y)
```

```python
c = model.intercept_
m = model.coef_
```

```python
print(c,m)
```

```python
print(f"y = {m[0]:.2f}x+{c:.2f}")
```

```python
y_pred = model.predict(x)
```

```python
import numpy as np
df['Predict'] = np.round(y_pred, 2)
```

```python
df
```

```python
MSE =  np.average((df['Price']-df['Predict'])**2)
```

```python
print(MSE)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
Mse = mean_squared_error(y,y_pred)
```

```python
print(Mse)
```

```python
r2 = r2_score(y,y_pred)
```

```python
print(r2)
```
