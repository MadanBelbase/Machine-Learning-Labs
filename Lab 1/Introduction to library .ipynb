```python
import numpy as np
```

```python
arr1 = np.array([10,20,30,40,50,60])
print("1-D Array:")
print(arr1)
```

```python
# Convert 1-D array to 2-D using reshape()
arr2 = arr1.reshape(2, 3)
```

```python
print("\n2-D Array after reshape:")
print(arr2)
```

```python
# Another array for addition
arr3 = np.array([[1, 2, 3],
                 [4, 5, 6]])

# Add two arrays
sum_array = arr2 + arr3

print("\nAddition of two arrays:")
print(sum_array)

# Shape of 2-D array
print("\nShape of 2-D array:")
print(arr2.shape)
```

```python
# 2. Access the last data of the 2-D array


print("\nLast element of 2-D array:")
print(arr2[-1, -1])
```

```python
# 3. Array Slicing
# 1-D slicing
print("\n1-D Array Slicing:")
print(arr1[1:4])

# Negative slicing
print("\nNegative Slicing:")
print(arr1[-4:-1])

# 2-D slicing
print("\n2-D Array Slicing:")
print(arr2[:, 1:3]) 
```

```python
# 4. Search data using where() and searchsorted()
```

```python
print("\nSearch using where():")
print(np.where(arr1 == 40))

sorted_arr = np.array([10, 20, 30, 40, 50])

print("\nSearch using searchsorted():")
print(np.searchsorted(sorted_arr, 35))
```

```python
# 5. Statistical Calculations
from scipy import stats
from sklearn.metrics import mean_squared_error
import math

data = np.array([12, 15, 18, 20, 22, 25, 30])

# Mean
mean = np.mean(data)

# Median
median = np.median(data)

# Mode
mode = stats.mode(data, keepdims=True)

# Standard Deviation
std_dev = np.std(data)

# Variance
variance = np.var(data)

print("\nStatistical Calculations")
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode.mode[0])
print("Standard Deviation:", std_dev)
print("Variance:", variance)

# MSE and RMSE
actual = np.array([3, -0.5, 2, 7])
predicted = np.array([2.5, 0.0, 2, 8])

mse = mean_squared_error(actual, predicted)
rmse = math.sqrt(mse)

print("\nMSE:", mse)
print("RMSE:", rmse)
```
