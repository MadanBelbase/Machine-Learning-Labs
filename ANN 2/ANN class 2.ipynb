```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python
# Create the dataset
data = {
    'Age': [36, 42, 23, 52, 43, 44, 66, 35, 52, 35, 24, 18, 45, 41],
    'Experience': [10, 12, 4, 4, 21, 14, 3, 14, 13, 5, 3, 3, 9, 5],
    'Rank': [9, 4, 6, 4, 8, 5, 7, 9, 7, 9, 5, 7, 9, 4],
    'Nationality': ['UK', 'USA', 'N', 'USA', 'USA', 'UK', 'N', 'UK', 'N', 'N', 'USA', 'UK', 'UK', 'N'],
    'Go': ['NO', 'NO', 'NO', 'YES', 'YES', 'NO', 'YES', 'YES', 'YES', 'YES', 'YES', 'YES', 'YES', 'NO']
}
```

```python
df = pd.DataFrame(data)

#Encode categorical variables
label_encoders= {}
for column in ['Nationality','Go']:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le
```

```python
#define features and target variables
x =  df.drop('Go',axis =1)
y = df['Go']

#split the dataset into training and testing 
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state =42)
```

```python
#convert the target variable to categorical 
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
print(y_train)

#create the neural network  model 
model = Sequential()
model.add(Dense(8,input_dim =x_train.shape[1],activation ='relu')) # input layer
model.add(Dense(8,activation ='tanh')) # hidden layer
model.add(Dense(2,activation ='sigmoid'))  # output layer (2 clasees)


```

```python
#Model Compilation
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

#Train the model
model.fit(x_train, y_train, epochs = 100, batch_size = 5,verbose = 1)
                                                                             
                                                                             

            
```

```python
#Evaluate the model

loss, accuracy = model.evaluate(x_test, y_test)
print(f'Test Accuracy: {accuracy: .2f}')

#Male Predictions
predictions = model.predict(x_test)
predicted_classes = predictions.argmax(axis = 1)
```

```python
# Convert predicted classes back to original labels
predicted_labels = label_encoders['Go'].inverse_transform(predicted_classes)
print(f'Predicted labels: {predicted_labels}')

from sklearn.metrics import confusion_matrix, classification_report

# Print confusion matrix and classification report
print(confusion_matrix(y_test.argmax(axis=1), predicted_classes))

print(classification_report(y_test.argmax(axis=1), predicted_classes))
```
