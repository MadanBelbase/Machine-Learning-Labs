```python
import numpy as np
 #generate a dummy data set

x = np.random.randint(10,50,100).reshape(20,5)
print(x.shape)
print(x)
#mean centring the data

x_mean = x-np.mean(x,axis=0)
print(x_mean)
```

```python
#calculate the co varinace matrix of the mean centered data

cov_mat = np.cov(x_mean, rowvar=False)
print(cov_mat)
```

```python
#calculating Eigen value and vectors 

eigen_value , eigen_vector = np.linalg.eigh(cov_mat)
print(eigen_value,eigen_vector)
```

```python
#sort the eigen value in desecending order 
sorted_index = np.argsort(eigen_value)[::-1]

sorted_eigen_value = eigen_value[sorted_index]

#similiarly sorth the vector

sorted_eigen_vector = eigen_vector[:,sorted_index]

print("Sorted Value", sorted_eigen_value)
print("Sorted vector", sorted_eigen_vector )
    
```

```python
#select the first n eigenvectors , n is desired dimension  of our final reduced data
n_comp = 2
eigen_vector_subset= sorted_eigen_vector[:,0:n_comp]
```

```python
#transform data 
x_reduced = np.dot(eigen_vector_subset.transpose(),x_mean.transpose()).transpose()

#visualize the shape
x_reduced.shape
```
